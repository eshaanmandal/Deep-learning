```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# defining the parameters of the line
x = 3
a = np.linspace(0,2,30)
a = a.reshape(-1,1)
b = x*a + np.random.randn(*a.shape)

# predicting 
U,S,Vt = np.linalg.svd(a,full_matrices= False)
x_t = Vt.T @ np.linalg.inv(np.diag(S)) @ U.T @ b
x_t

# curve fitting
y_pred = x_t*a

# plot the data
plt.plot(a,x*a,label='original')
plt.plot(a,b,'ob',label='Gaussian Noise')
plt.plot(a,y_pred,'--r',label='Predicted')

plt.legend()
plt.show()

# Original slope, and predicted slope
print(f'Original slope :{x}\nPredicted slope :{x_t}')
```
